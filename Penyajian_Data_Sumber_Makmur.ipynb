# **Deskripsi Dataset**

Data ini berisikan 173 mahasiswa UPN “Veteran” Jawa Timur untuk mengetahui kebiasaan mereka di kantin kampus. Di dalamnya tercatat fakultas, prodi, serta NPM responden, lalu seberapa banyak uang jajan yang mereka habiskan, berapa kali mereka makan tiap minggu, dan berapa lama biasanya menunggu pesanan. Mahasiswa juga memberi nilai 1–5 untuk rasa, harga, kenyamanan, rekomendasi, dan kemudahan bayar QRIS. Setelah data dibersihkan—duplikat dihapus, tipe angka diseragamkan, nilai kosong diisi, serta angka negatif disesuaikan—kami menambahkan dua label praktis: kategori pengeluaran (rendah, sedang, tinggi) dan kategori frekuensi makan (jarang, cukup sering, sangat sering). Hasil akhirnya adalah 17 kolom yang rapi tanpa celah, mudah dipahami, dan data siap untuk ke tahap selanjutnya yaitu penyajian data.

### **PIE CHART**

```python
# 2. PROPORSI FREKUENSI MAKAN (PIE CHART) ---

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat kanvas gambar berukuran 10x10 inci agar pie chart terlihat bulat sempurna
plt.figure(figsize=(6, 6))

# Menghitung jumlah data untuk setiap kategori frekuensi makan
frekuensi_counts = df['kategori_frekuensi_makan'].value_counts()

# Membuat pie chart
plt.pie(
    frekuensi_counts,                                   # Data jumlah untuk setiap kategori
    labels=frekuensi_counts.index,                      # Label untuk setiap potongan pie
    autopct='%1.1f%%',                                  # Format untuk menampilkan persentase
    startangle=140,                                     # Sudut awal untuk potongan pertama
    colors=sns.color_palette('pastel')                  # Skema warna yang lembut
)

# Menambahkan judul
plt.title('Proporsi Frekuensi Makan per Minggu', fontsize=18, fontweight='bold')
plt.ylabel('') # Menghilangkan label 'ylabel' yang tidak perlu

# Menampilkan plot
plt.show()
```

Grafik diatas merupakan Pie Chart yang menampilkan Proporsi Frekuensi Makan di Kantin per Minggu. Alasan menggunakan Pie Chart adalah karena ingin mengetahui distribusi bagian-bagian dari total keseluruhan dalam data frekuensi makan di kantin dalam satu minggu yang berbentuk persentase. Berdasarkan Pie Chart diatas dapat diketahui bahwa:

*   Sangat Sering (>4x) merupakan kategori frekuensi yang memiliki hasil paling rendah yaitu 7.5%
*   Cukup Sering (3-4x) merupakan kategori frekuensi yang memiliki hasil 39.3%
*   Jarang (1-2x) merupakan kategori frekuensi yang memiliki hasil terbesar yaitu 53.2%

Berdasarkan hasil tersebut dapat ditarik kesimpulan bahwa mayoritas mahasiswa jarang pergi ke kantin dalam 1 minggunya hal ini dapat dilihat dari jumlah persentase pada Pie Chart sebesar 53.2% sedangkan hanya 7.5% responden saja yang sangat sering ke kantin dalam 1 minggunya dan sisanya cukup sering dengan persentase 39.3%



```python
# 3. PIE CHART
# Memvisualisasikan proporsi kategori pengeluaran mahasiswa.
pengeluaran_count = df['kategori_pengeluaran'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(pengeluaran_count,
        labels=pengeluaran_count.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=sns.color_palette("pastel"))
plt.title('Proporsi Kategori Pengeluaran Mahasiswa')
plt.show()
```

Grafik diatas merupakan Pie Chart yang menampilkan Proporsi Kategori Pengeluaran Mahasiswa dalam satu kali makan di kantin UPN. Alasan menggunakan Pie Chart adalah karena ingin mengatahui distribusi pengeluaran mahasiswa dalam satu kali makan di kantin UPN yang berbentuk persentase. Pada Bar Chart terdapat 3 kategori pengeluaran diataranya:
*   Rendah merupakan kategori pengeluaran kurang dari Rp 15.000 dengan nilai 48.6%.
*   Sedang merupakan kategori pengeluaran berkisar diatas Rp 15.000 - Rp 25.000 dengan nilai 41.6%.

Berdasarkan hasil Pie Chart dapat diketahui bahwa kategori rendah dan sedang memiliki persentase yang selisihnya tidak jauh beda meskipun kategori rendah masih menjadi persentase yang paling tinggi nilainya dengan 48.6%. ### **BOXPLOT**

```python
# 5. BOXPLOT
# Membandingkan kenyamanan kantin berdasarkan kategori pengeluaran.
plt.figure(figsize=(7,4))
sns.boxplot(x='kategori_pengeluaran',
            y='skala_kenyamanan(1-5)',
            data=df,
            palette='Set2')
plt.title('Sebaran Kenyamanan Berdasarkan Kategori Pengeluaran')
plt.xlabel('Kategori Pengeluaran')
plt.ylabel('Skala Kenyamanan (1-5)')
plt.show()
```

Grafik diatas merupakan sebuah Boxplot sebaran kenyamanan berdasarkan kategori pengeluaran. Alasan menggunakan boxplot adalah untuk mengetahui sebaran rentang kenyamanan berdasarkan kategori pengeluaran. Berdasarkan boxplot diatas dapat ditarik informasi:
* Median yang Sama: Garis tengah di dalam setiap kotak berada pada posisi yang sama untuk ketiga kategori, yaitu pada skala kenyamanan 2.5. Ini adalah temuan kunci yang menunjukkan bahwa tidak ada perbedaan tingkat kenyamanan secara rata-rata antara kelompok kategori pengeluaran rendah, sedang dan tinggi.
* Range yang Mirip: "Jangkauan"(whiskers) dari boxplot juga hampir identik, membentang dari nilai minimum 1.0 hingga nilai maksimum sekitar 4.0 - 4.5. Ini menunjukkan bahwa rentang keseluruhan pengalaman kenyamanan tidak jauh berbeda antar kelompok.
* Sebaran Data yang Identik: Kotak (box) pada ketiga plot, yang mencakup 50% data tengah (Interquartile Range/IQR), memiliki rentang yang sama, yaitu dari skala 2.0 (kuartil bawah/Q1) hingga 3.0 (kuartil atas/Q3). Ini berarti, variabilitas tingkat kenyamanan untuk mayoritas responden di setiap kelompok adalah sama.

### **STACKED BAR CHART**

```python
# 6. KOMPOSISI PENGELUARAN (STACKED BAR CHART)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-talk')
url = 'https://docs.google.com/spreadsheets/d/1AFV6MwL3kkeqPZG1NLJ3T__vebWK4pQtBeD6He0nSn4/export?format=csv&gid=215109113'
df = pd.read_csv(url)

# Menghitung tabulasi silang dan mengubahnya menjadi persentase
komposisi = pd.crosstab(df['Fakultas'], df['kategori_pengeluaran'], normalize='index') * 100

# Membuat stacked bar chart horizontal
komposisi.plot(
    kind='barh',
    stacked=True,
    figsize=(14, 10),
    colormap='viridis'
)

# Menambahkan judul dan label
plt.title('Komposisi Kategori Pengeluaran di Setiap Fakultas', fontsize=18, fontweight='bold')
plt.xlabel('Persentase (%)', fontsize=14)
plt.ylabel('Fakultas', fontsize=14)
plt.legend(title='Kategori Pengeluaran', bbox_to_anchor=(1.05, 1), loc='upper left') # Memindahkan legenda keluar
plt.tight_layout()
plt.show()
```

Grafik diatas merupakan Stacked Bar Chart tentang komposisi kategori pengeluaran di setiap fakultas. Alasan menggunakan stacked bar chart adalah ingin mengetahui persentase komposisi kategori pengeluaran di setiap fakultas. Berdasarkan stacked bar chart dapat ditarik informasi:
* Fakultas dengan pengeluaran dominan rendah berada di Fakultas Ilmu Sosial dan Ilmu Politik, FP dan FK dengan persentase 100%
* Fakultas dengan pengeluaran dominan sedang berada di Fakultas Ilmu Sosial dan Budaya dan Administrasi Publik dengan persentase 100%
* Fakultas dengan pengeluaran dominan tinggi berada di FTS dan FEB dengan persentase 20%
* Fakultas dengan komposisi paling beragam berada di FTS dan FEB karena Di kedua fakultas ini, tidak ada satu kategori pun yang mendominasi secara absolut. Proporsinya terbagi cukup merata antara "Rendah", "Sedang" , dan "Tinggi".

### **SCATTER PLOT**

```python
# 7. POLA BELANJA VS WAKTU TUNGGU PER FAKULTAS (SCATTER PLOT)

# Import library yang dibutuhkan
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Atur gaya visual
plt.style.use('seaborn-v0_8-talk')

# Baca data dari link Google Sheets
url = 'https://docs.google.com/spreadsheets/d/1AFV6MwL3kkeqPZG1NLJ3T__vebWK4pQtBeD6He0nSn4/export?format=csv&gid=215109113'
df = pd.read_csv(url)

# 1. MEMBUAT VISUALISASI
# Membuat kanvas gambar berukuran 14x9 inci
plt.figure(figsize=(14, 9))

# Membuat scatter plot murni menggunakan seaborn.scatterplot
sns.scatterplot(
    data=df,
    x='waktu_tunggu_menit',  # Variabel numerik pertama di sumbu x
    y='rata_rata_uang',      # Variabel numerik kedua di sumbu y
    hue='Fakultas',          # 'hue' akan memberi warna berbeda untuk setiap fakultas
    palette='viridis',       # Palet warna untuk membedakan fakultas
    alpha=0.7,               # 'alpha' mengatur transparansi titik agar tumpukan terlihat
    s=100                    # 's' mengatur ukuran titik agar lebih jelas terlihat
)

# 2. MENAMBAHKAN JUDUL DAN LABEL
plt.title('Pola Perilaku Belanja vs. Waktu Tunggu per Fakultas', fontsize=18, fontweight='bold')
plt.xlabel('Waktu Tunggu (Menit)', fontsize=14)
plt.ylabel('Rata-Rata Uang yang Dihabiskan (Rp)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6) # Menambahkan grid untuk membantu pembacaan
plt.legend(title='Fakultas', bbox_to_anchor=(1.05, 1), loc='upper left') # Memindahkan legenda ke luar plot

# 3. Merapikan layout dan menampilkan plot
plt.tight_layout()
plt.show()
```

Grafik diatas merupakan Scatter Plot tentang pola perilaku belanja vs waktu tunggu per fakultas. Alasan menggunakan scatter plot adalah untuk menunjukkan hubungan atau sebuah pola tentang rata-rata uang yang dihabiskan dengan waktu tunggu(menit).

* Sumbu X mewakili waktu tunggu yang mencakup rentang dari 0 menit hingga 30 menit.
* Sumbu Y mewakili rentang uang yang dihabiskan dari Rp 7.500 - Rp 27.500
* Berdasarkan scatter plot tidak ada korelasi yang jelas antara waktu tunggu dan jumlah belanja. Sebaran titik data tidak menunjukkan pola linear yang jelas baik menanjak maupun menurun. Ini berarti tidak ada hubungan yang kuat antara lamanya seseorang menunggu dengan jumlah uang yang mereka habiskan.
* Berdasarkan pola dari scatter plot tidak ada pola spesifik. Namun dapat diketahui bahwa mayoritas waktu tunggu relatif singkat dimana terdapat konsentrasi titik data yang sangat padat di sisi kiri grafik, terutama pada rentang waktu tunggu 0 hingga 10 menit.
* Berdasarkan pola dari scatter plot dapat diketahui pola pengeluaran mengelompok pada nominal tertentu. Titik-titik data cenderung membentuk garis-garis horizontal yang padat pada nominal-nominal tertentu, seperti Rp 10.000, Rp 15.000, Rp 20.000, dan Rp 25.000.

### **HEATMAP**

```python
# 8. ANALISIS KORELASI (HEATMAP)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Memilih hanya kolom-kolom skala untuk dianalisis
kolom_skala = [
    'skala_kenyamanan(1-5)', 'skala_harga(1-5)', 'skala_rasa(1-5)',
    'skala_qris(1-5)', 'skala_rekomendasi(1-5)', 'skala_antrean(1-5)'
]
matriks_korelasi = df[kolom_skala].corr()

# Membuat kanvas gambar berukuran 12x10 inci
plt.figure(figsize=(12, 10))

# Membuat heatmap
sns.heatmap(
    matriks_korelasi, # Data matriks korelasi
    annot=True,       # Menampilkan angka di dalam setiap kotak
    cmap='coolwarm',  # Skema warna, merah (positif kuat), biru (negatif kuat)
    fmt='.2f',        # Format angka menjadi 2 desimal
    linewidths=.5     # Memberi garis pemisah antar kotak
)

# Menambahkan judul dan memutar label agar mudah dibaca
plt.title('Heatmap Korelasi Antar Faktor Kepuasan', fontsize=18, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Merapikan dan menampilkan plot
plt.tight_layout()
plt.show()
```

Heatmap diatas merupakan heatmap tentang korelasi antar faktor kepuasan dimana terdapat 6 faktor kepuasan yaitu kenyamanan, harga, rasa, qris, rekomendasi dan antrean. Alasan menggunakan heatmap adalah untuk menganalisa korelasi atau hubungan antar faktor kepuasan. Berdasarkan heatmap diatas dapat ditarik informasi:
* Korelasi Positif: Korelasi positif kuat apabila mendekati 1. Jika satu faktor naik, faktor lainnya cenderung ikut naik. Misalnya 'skala_rekomendasi' dan 'skala_kenyamanan' dengan nilai 0.53 hal tersebut menandakan bahwa korelasi cekup kuat dimana semakin tinggi tingkat kenyamanan kantin maka semakin tinggi juga seseorang akan merekomendasikan kantin UPN.
* Korelasi Negatif: Korelasi negatif kuat apabila mendekati -1. Jika satu faktor naik, faktor lainnya cenderung turun. Misalnya 'skala_antrean' dan 'skala_rekomendasi' dengan nilai -0.14 hal tersebut menandakan bahwa korelasi negatif yang cukup lemah dimana kecil hubungannya apabila semakin tinggi antrean maka terjadi penurunan seseorang akan merekomendasikan kantin UPN.
* Nilai Korelasi: Semakin dekat nilai ke 1 atau -1, semakin kuat korelasinya. Nilai mendekati 0 menunjukkan korelasi yang lemah atau tidak ada korelasi. Misalnya 'skala_qris' dan 'skala_antrean' yang memiliki nilai 0.02 yang dimana itu dekat dengan 0 yang dimana hubungan sangat lemah atau tidak ada sama sekali.

### **LOLLIPOP CHART**

```python
# 9. LOLLIPOP CHART
# Fungsi: Menampilkan rata-rata tiap aspek kepuasan secara rapi dan modern.
# Mudah membandingkan angka rata-rata antar aspek karena ada titik & garis sebagai indikator.

aspek = [
    'skala_kenyamanan(1-5)',
    'skala_harga(1-5)',
    'skala_rasa(1-5)',
    'skala_qris(1-5)',
    'skala_rekomendasi(1-5)',
    'skala_antrean(1-5)'
]
# Hitung rata-rata tiap aspek
mean_aspek = df[aspek].mean()

# Plot Lollipop Chart
plt.figure(figsize=(9,5))
(markerline, stemlines, baseline) = plt.stem(
    mean_aspek.index,        # nama aspek
    mean_aspek.values,       # nilai rata-rata
    basefmt=" ",             # hilangkan garis dasar
    linefmt='grey',          # warna garis
    markerfmt='o'            # bentuk titik
)

# Set warna titik
plt.setp(markerline, markersize=8, markerfacecolor='tab:blue')

# Judul dan label sumbu
plt.title('Lollipop Chart Rata-rata Aspek Kepuasan Mahasiswa', fontsize=14)
plt.ylabel('Skor Rata-rata (1 = sangat buruk, 5 = sangat baik)')
plt.xticks(rotation=30)
plt.ylim(0, 5.5)  # skala kepuasan 1-5

# Tambahkan nilai di ujung garis
for i, val in enumerate(mean_aspek.values):
    plt.text(i, val + 0.1, f"{val:.2f}", ha='center', va='bottom', fontsize=10, color='black')

plt.tight_layout()
plt.show()
```

Grafik diatas merupakan diagram Lollipop Chart yang menampilkan rata-rata aspek kepuasan mahasiwa berdasarkan beberapa kategori yaitu kenyamanan, harga, rasa, qris, rekomendasi dan antrean. Alasan menggunakan Lollipop Chart adalah ingin menyajikan data yang memiliki banyak kategori tetapi ingin terlihat sederhana secara visual sehingga bisa lebih mudah dipahami. Berdasarkan Lollipop Chart tersebut dapat diketahui bahwa:
* skala_kenyamanan(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 2.48
* skala_harga(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 3.32
* skala_rasa(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 3.66
* skala_qris(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 4.12
* skala_rekomendasi(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 2.50
* skala_antrean(1-5) memiliki nilai rata-rata aspek kepuasan sebesar 3.90

Berdasarkan hasil Lollipop Chart tersebut dapat disimpulkan bahwa aspek kepuasan mahasiswa terhadap kantin UPN sudah cukup baik hal itu dapat dilihat dari rata-rata 4 kategori sudah lebih dari 3. Namun terdapat masalah utama pada kategori skala_kenyamanan dan skala_rekomendasi yang nilainya dibawah 3 yang menandakan bahwa para responden rata-rata kurang puas terhadap aspek tersebut.
